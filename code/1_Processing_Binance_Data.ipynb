```python
import pandas as pd
pd.set_option("display.max_columns", None)

df_binance = pd.read_csv('../data/raw/Binance_Data_1m_2021_2024.csv')
df_binance.head()
```

```python
df_binance['open_time_dt'] = pd.to_datetime(df_binance['open_time'], unit='ms', utc=True)
df_binance['minute'] = df_binance['open_time_dt'].dt.floor('min')
df_binance['date'] = df_binance['open_time_dt'].dt.date

df_binance['benchmark_close'] = df_binance['close']
```

```python
df_binance.to_csv('../data/processed/binance_1m_processed.csv', index=False)
```
