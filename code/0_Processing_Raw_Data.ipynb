```python
import pandas as pd
pd.set_option("display.max_columns", None)

df = pd.read_csv('../data/raw/bq-results-raw-full.csv')

df.rename(columns={
    'topics0': 'topic0',
    'topics1': 'topic1',
    'topics2': 'topic2',
    'topics3': 'topic3'
}, inplace=True)

df['block_timestamp'] = pd.to_datetime(df["block_timestamp"], errors="coerce", utc=True)
df['address'] = df['address'].str.lower()
df['topic0'] = df['topic0'].str.lower()
df['topic1'] = df['topic1'].str.lower()
df['topic2'] = df['topic2'].str.lower()
df['topic3'] = df['topic3'].str.lower()

df.head()
```
